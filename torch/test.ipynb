```python
import torch
```

```python
h=3
token_len = 6
d_model = 6
dv = dk = d_model//h
input_shape = (token_len, d_model)
input = torch.rand(input_shape)
wq = []
for _ in range(h):
    wq.append(torch.rand((d_model, dk)))

WQ = torch.cat(wq, dim = 1)

results = []
for i in range(h):
    results.append(input @ wq[i])

Q = input @ WQ

results = torch.cat(results, dim=1)

print(Q)
print()
print(results)
print()
print(Q==results)
```
